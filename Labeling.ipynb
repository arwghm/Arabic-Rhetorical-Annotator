```python

!pip install python-docx -q

import docx
import pandas as pd

document = docx.Document('bc4.docx')
sentences = [p.text.strip() for p in document.paragraphs if len(p.text.strip()) > 10]
df = pd.DataFrame({'الجملة': sentences})
print(f"✅ عدد الجمل: {len(df)}")
```

```python

تشبيه_كلمات   = ['كأن', 'كأنه', 'كأنها', 'مثل', 'يشبه', 'كال', 'كـ', 'شبيه']
استعارة_كلمات = ['يبكي', 'يضحك', 'يصرخ', 'يتألم', 'الليل', 'الزمان', 'الدهر']
كناية_كلمات   = ['كثير الرماد', 'طويل النجاد', 'نائم الضحى', 'ثقيل الظل']

def detect(line):
    if any(w in line for w in تشبيه_كلمات):   return 'تشبيه'
    if any(w in line for w in استعارة_كلمات): return 'استعارة'
    if any(w in line for w in كناية_كلمات):   return 'كناية'
    return None

results = []
for line in df['الجملة']:
    label = detect(line)
    if label:
        results.append({'الجملة': line, 'النوع المقترح': label})

df_filtered = pd.DataFrame(results)
print(f" الجمل المرشحة: {len(df_filtered)}")
print(df_filtered.head(100))
```

```python
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', 100)
print(df_filtered.head(100))
```

```python

import ipywidgets as widgets
from IPython.display import clear_output, display

دفعة = 0  # ← غيري من 0 إلى 9 كل مرة

start = دفعة * 10
batch = df_filtered.iloc[start:start+10]
labels_batch = []
current = [0]

btn1 = widgets.Button(description='تشبيه',   button_style='success')
btn2 = widgets.Button(description='استعارة', button_style='info')
btn3 = widgets.Button(description='كناية',   button_style='warning')
btn0 = widgets.Button(description='تخطي',    button_style='danger')

def show():
    clear_output(wait=True)
    i = current[0]
    print(f"الجملة {start+i+1} | المقترح: {batch.iloc[i]['النوع المقترح']}")
    print("─" * 40)
    print(f"\n{batch.iloc[i]['الجملة']}\n")
    display(widgets.HBox([btn1, btn2, btn3, btn0]))

def click(label):
    labels_batch.append(label)
    current[0] += 1
    if current[0] < 10:
        show()
    else:
        pd.DataFrame({'الجملة': batch['الجملة'].tolist(), 'التصنيف': labels_batch}).to_csv(f'batch_{دفعة}.csv', encoding='utf-8-sig', index=False)
        clear_output(wait=True)
        print(f"✅ تم حفظ الدفعة {دفعة+1}")

btn1.on_click(lambda x: click('تشبيه'))
btn2.on_click(lambda x: click('استعارة'))
btn3.on_click(lambda x: click('كناية'))
btn0.on_click(lambda x: click('غير محدد'))

show()
```
